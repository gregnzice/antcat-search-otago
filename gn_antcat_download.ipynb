## Import the download function

```python
import requests

url = "https://raw.githubusercontent.com/antarcticanz/antcat-search/main/gn_antcat_download.py"
with open("gn_antcat_download.py", "wb") as f:
    f.write(requests.get(url).content)
```

```python
from gn_antcat_download import gn_antcat_download

records, drill_hole_data = gn_antcat_download(
    search_term="DRILL HOLE",
    filter_terms=["ICE DEPTH/THICKNES", "SUB-ICE PLATELET LAYER"]
)
```

## Inspect the data

```python
import pandas as pd
pd.set_option("display.max_colwidth", None)  # show full column contents

drill_hole_data
```

## Plot/Map the data

```python
import folium

m = folium.Map(location=[-77.55, 166.4], zoom_start=7, tiles='Esri WorldImagery')

for _, row in drill_hole_data.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=5,
        color='red',
        fill=True,
        fill_color='blue',
        fill_opacity=0.7
    ).add_to(m)

m
```
